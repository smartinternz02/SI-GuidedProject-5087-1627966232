```python
x = ds.iloc[:, :-1].values
y = ds.iloc[:, 16].values
```

```python
ct = ColumnTransformer([("one",OneHotEncoder(),[1,2,3,4,6,7,8,10,15])],remainder = "passthrough")
x = ct.fit_transform(x)
```

```python
x
```

```python
y = ds.deposit.replace(to_replace=['no', 'yes'], value=[0, 1])
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
reg=Sequential()
reg.add(Dense(units=17,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=34,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=34,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=34,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=34,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=34,kernel_initializer= "random_uniform",activation= "relu"))
reg.add(Dense(units=1,kernel_initializer= "random_uniform",activation= "relu"))
reg.compile(optimizer="rmsprop",loss="mean_squared_error",metrics=["accuracy"])
```

```python
reg.fit(x_train,y_train,epochs=100)
```

```python
ypred = reg.predict(x_test)
ypred
```
